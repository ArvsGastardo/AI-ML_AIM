## I. Background and Motivation ##

A startup company, currently having a project with The United Nations Development Programme (UNDP) and the Ormoc City local government unit (LGU),  intends to determine what design aspects of the vertical wind turbine (VWT) need to be prioritized to fit customer preferences, aligning with the circular economy (EC) and renewable energy (RE) of the organization. The team initially identified social, economic, environmental, technical performance, installation and infrastructure, Policy & regulations, and user experience & market acceptance as significant design factors. Neutrosophic Logic–Based was used, assigning Truth (T), Indeterminacy (I), and False (F) with values 0.70, 0.20, and 0.10, respectively. Although the survey results showed interest in the majority of the aspects set, it revealed that social design factors, technical performance factors, user experience & market acceptance factors are the top three of the customer’s interests.
